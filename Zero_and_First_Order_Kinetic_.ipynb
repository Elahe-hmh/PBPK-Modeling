# DETERMINISTIC


In every step, I transfer a fraction of island 1 to island 2 and a fraction of island 2 to island 1.

```python
def func(p10, p20,a,b, n):
  p1_listd=[p10]
  p2_listd=[p20]
  p1,p2=p10,p20
  p1f=p10
  p2f=p20
  for i in range(n):
    p1f=p1
    p2f=p2
 #   print(p1f,p2f)
    p1 = p1 + a*p2f - b*p1f
    p2 = p2 + b*p1f - a*p2f

#    print(p1,p2)
    
    p1_listd.append(p1)
    p2_listd.append(p2)
 #   print(p1+p2)
  return(p1_listd,p2_listd)
```

```python
(p1_listd , p2_listd)=func(4000,10000,0.8,0.3,20)
```

```python
plt.plot(p1_listd)
```

```python
plt.plot(p2_listd)
```

```python
plt.plot(p1_listd)
plt.plot(p2_listd)

plt.legend(["population 1", "population 2"])
```

We can see the stability rate that the populations in two islands get to, is so fast cause we are moving the whole passenger at one time.

## STOCHASTIC

Here we look at the problem in a probabilistic way.

---
Every person can go to the other island with a probalility


```python
from random import seed
from random import random
```

```python
def funcstoc(p10,p20,a,b,n, alpha):
  p1_lists=[p10]
  p2_lists=[p20]
  (p1,p2)=(p10,p20)

  for l in range(n):
    n1=p10
    n2=p20
    p1_move=0
    p2_move=0
    
    for j in range(int(p1*alpha)):
      value=random()
      if  value <= b and p1!=0:
        p2_move+= 1
    for k in range(int(p2*alpha)):
      value=random()
      if value <= a and a!=0 and p2!=0:
        p1_move+=1
    p1=p1+p1_move-p2_move
    p2=p2+p2_move-p1_move
    
    p1_lists.append(p1)
    p2_lists.append(p2)  
  return(p1_lists,p2_lists)  
```

## alpha=1
I go to **every person** 
 
give them a random number from 0 to 1

if their number is less than "a" in schematic image, they can move. if not they will stay at home =)

What I'm doing? I'm investigating all of the population in island 1 at once.

so first: it's gonna be so fast 

secondly, I'm considering the population of each island. If one has 4000 people and the other one has 10000, I will spend more time on the second one and will move more people from there

## ==> FIRST ORDER KINETIC

###    dN1/dt= - aN1 + bN2

```python
p1_lists,p2_lists=funcstoc(4000,10000,0.8,0.3,100,1)

plt.plot(p1_lists)
plt.plot(p2_lists)

plt.legend(["population 1", "population 2"])
```

# It's ZERO ORDER KINETIC

## dN1/dt= - a + b =100 =))


```python
def funcstoc(p10,p20,a,b,n, nperson):
  p1_lists=[p10]
  p2_lists=[p20]
  (p1,p2)=(p10,p20)

  for l in range(n):
    n1=p10
    n2=p20
    p1_move=0
    p2_move=0
    
    for j in range(nperson):
      value=random()
      if  value <= b and p1!=0:
        p2_move+= 1
    for k in range(nperson):
      value=random()
      if value <= a and a!=0 and p2!=0:
        p1_move+=1
    p1=p1+p1_move-p2_move
    p2=p2+p2_move-p1_move
    
    p1_lists.append(p1)
    p2_lists.append(p2)  
  return(p1_lists,p2_lists)  
```

```python
p1_lists,p2_lists=funcstoc(4000,10000,0.8,0.3,100,100)

plt.plot(p1_lists)
plt.plot(p2_lists)

plt.legend(["population 1", "population 2"])
```
