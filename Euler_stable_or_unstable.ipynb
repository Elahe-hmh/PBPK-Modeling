```python
from matplotlib import pyplot as plt
```

```python
x0=1
v0=0
```

```python
x=x0
v=v0
xlist=[x0]
deltat=0.1
xfix=x
for i in range(500):
  v= v +-2 * xfix * deltat
  vfix=v
  x= x + vfix * deltat
  xfix=x
  xlist.append(x)
  
```

```python
plt.plot(xlist)
```
